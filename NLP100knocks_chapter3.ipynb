## 21. カテゴリ名を含む行を抽出
記事中でカテゴリ名を宣言している行を抽出せよ．

```python
import codecs
import json
import re


def nlpknock(title_word):
    for row in codecs.open("./datasets/jawiki-country.json", "r", "utf-8"):
        article = json.loads(row)
        if title_word == article["title"]:
            return article["text"]

word = nlpknock(u"イギリス")
for line in word.split("\n"):
    if re.search(r"Category:", line):
        print(line)
```

## 22. カテゴリ名の抽出
記事のカテゴリ名を（行単位ではなく名前で）抽出せよ．

```python
#問22の出力から名前だけ抽出する正規表現を加える。
for name in word.split("\n"):
    matched = re.search(r"Category:(?P<category>.+?)(\||])", name)
    if matched:
        print(matched.group("category"))
```

## 23. セクション構造
記事中に含まれるセクション名とそのレベル（例えば"== セクション名 =="なら1）を表示せよ．

wikipediaのMarkdown?を参照すると「#」的な役割らしいhttps://ja.wikipedia.org/wiki/Help:早見表

\\n=== ムハンマド・アリー朝 ===\\n　のような記法

```python
import re

word = nlpknock(u"イギリス")

#改行区切り（\n）したものを一行づつ正規表現に従って検索。
for line in word.split("\n"):
    #シンボリックグループ<level>と<secname>を付けて全文検索
    result = re.search(r"^(?P<level>=+)(?P<secname>.+)\1$", line)
    
    #論理式Trueのみ処理でlevelとsecnameを表示
    if result:
        level = result.group("level").count("=") -1
        secname = result.group("secname")
        print("{0}: {1}".format(level, secname))
```

## 24. ファイル参照の抽出
記事から参照されているメディアファイルをすべて抜き出せ．
こんなかたち↓

\\n|ファイル:Egypt.Giza.Sphinx.01.jpg|[[メンフィスとその墓地遺跡|メンフィスとその墓地遺跡-ギーザからダハシュールまでのピラミッド地帯]] - (1979年、文化遺産)

```python
import re

word = nlpknock(u"イギリス")

#改行区切り（\n）したものを一行づつ正規表現に従って検索。
for line in word.split("\n"):
    #シンボリックグループ<mediafile>を付けて全文検索
    result = re.search("ファイル:(?P<mediafile>[^|]+)\|", line)
    
    #論理式Trueのみ処理でlevelとsecnameを表示
    if result:
        print(result.group("mediafile"))
```

## 25. テンプレートの抽出
記事中に含まれる「基礎情報」テンプレートのフィールド名と値を抽出し，辞書オブジェクトとして格納せよ．
（問20を確認して「基礎情報」の構造を確認）

```python
from pprint import pprint

#{{基礎情報 ~~ }}な部分だけ抽出する関数
def nlpknock_kiso(word):
    result = re.search("{{基礎情報[^|]+\|(?P<field_value>.+?)\n}}", word, re.DOTALL)
    if not result:
        return {}
    
    field_value = result.group("field_value")
    
    #
    fv_dict ={}
    #スペースとイコールでfieldとvalueに分割
    for item in field_value.split("\n|"):
        field, value = re.split(r"\s+=\s+", item, maxsplit=1)
        fv_dict[field] = value
        
    return fv_dict
    
word = nlpknock(u"イギリス")
kiso = nlpknock_kiso(word)

pprint(kiso, indent=4)
```

## 26. 強調マークアップの除去
25の処理時に，テンプレートの値からMediaWikiの強調マークアップ（弱い強調，強調，強い強調のすべて）を除去してテキストに変換せよ  
（参考: マークアップ早見表：https://ja.wikipedia.org/wiki/Help:早見表 ）

'''強調'''  
確立形態4': "現在の国号「'''グレートブリテン及び北アイルランド連合王国'''」に変更"

```python
def rm_emphasis(word):
    #2回以上"'"が使われる部分は全部を""で空白に
    return re.sub(r"'{2,}", "",word)

word = nlpknock(u"イギリス")
kiso = nlpknock_kiso(word)

rmph_kiso = {}
for field, value in kiso.items():
    value = rm_emphasis(value)
    rmph_kiso[field] = value
    
pprint(rmph_kiso,indent=4)
```

## 27. 内部リンクの除去
26の処理に加えて，テンプレートの値からMediaWikiの内部リンクマークアップを除去し，テキストに変換せよ（参考: マークアップ早見表）

[[記事名]]
[[記事名|表示文字]]
[[記事名#節名|表示文字]]

```python
def rm_internallink(word):
    return re.sub(r"\[\[([^]]+)\]\]", lambda m: m.group(1).split("|")[-1], word)

word = nlpknock(u"イギリス")
kiso = nlpknock_kiso(word)

rmph_kiso = {}
for field, value in kiso.items():
    value = rm_emphasis(value)
    value = rm_internallink(value)
    rmph_kiso[field] = value
    
pprint(rmph_kiso,indent=4)
```
